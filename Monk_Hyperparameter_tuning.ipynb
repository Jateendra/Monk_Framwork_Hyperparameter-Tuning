```python
from monk.gluon_prototype import prototype
```

```python
gtf = prototype(verbose=1);
gtf.Prototype("select_optimizer", "default_sgd");
```

```python
# gtf.Default(dataset_path="/content/training/training",
gtf.Default(dataset_path=["/content/training/training", "/content/validation/validation"], 
            model_name="resnet18_v1", 
            freeze_base_network=True,
            num_epochs=5);
```

```python
gtf.List_Optimizers();
```

```python
# Reset train and validation transforms
gtf.optimizer_adam(0.001);


# Very important to Reset Transforms post 
gtf.Reload();
```

```python
gtf.lr_fixed();
```

```python
gtf.loss_softmax_crossentropy()
```

```python
#Start Training
gtf.Train();
```

```python
# Create Project and Experiment

gtf = prototype(verbose=1);
gtf.Prototype("select_optimizer", "default_sgd", eval_infer=True);
```

```python
gtf.Dataset_Params(dataset_path="/content/validation/validation");
gtf.Dataset();
```

```python
accuracy, class_based_accuracy = gtf.Evaluate();
```
